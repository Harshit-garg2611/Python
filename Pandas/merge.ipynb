```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id':[101,102,103,106,108,110],
    'name':['Harshit', 'Aarti', 'Mehak', 'Yashika', 'Mohit', 'Udemy']})


df2 = pd.DataFrame({
    'id':[101,102,104,106,108,112,114], 
    'course_name':['DA', 'CIVIL', 'BEAUTY', 'IAS', 'TRADING', 'HACKING','DSA']})

```

```python
# pd.merge(left,right,how,on)

pd.merge(df1,df2,how='inner',on='id')
```

```python

pd.merge(df1,df2,how='left',on='id')

# inner,lefy,right,outer,cross
```

```python
# if we have common column with different names than use left_on and right_on to merge 
df3 = pd.DataFrame({
    'id1':[101,102,103,106,108,110],
    'name':['Harshit', 'Aarti', 'Mehak', 'Yashika', 'Mohit', 'Udemy']})


df4 = pd.DataFrame({
    'id2':[101,102,104,106,108,112,114], 
    'course_name':['DA', 'CIVIL', 'BEAUTY', 'IAS', 'TRADING', 'HACKING','DSA']})
```

```python
pd.merge(df3,df4,how='outer',left_on='id1',right_on='id2')
```

## Group by :
- DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)

> Parameters :
 
- by: Required parameter to specify the column(s) to group by.
- axis: Optional, specifies the axis to group by (default is 0 for rows).
- level: Optional, used for grouping by a certain level in a MultiIndex.
- as_index: Optional, whether to use the group labels as the index (default is True).
- sort: Optional, whether to sort the group keys (default is True).
- group_keys: Optional, whether to add the group keys to the index (default is True).
- dropna: Optional, whether to include rows/columns with NULL values (default is True)
