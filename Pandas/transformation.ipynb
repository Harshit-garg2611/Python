```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/Netflix.csv')
```

```python
df.head()
```

```python
# make release year 2 year less for all the shows (.apply function)

def pre_release_shows(x):
    return x-2 

data = df['release_year'].apply(pre_release_shows)

data.head()


#  remember apply does not change the original data frame
```

```python
df.head()
```

```python
df['release_year'] = df['release_year'].apply(pre_release_shows)

df.head()
```

```python
# instead of writing new function we can define lamda function

# Lambda function 

# lambda x : x-2
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format='mixed',utc=False)

df.head()
```

```python
# apply lamda function pre covid , post covid release 

df['release_year'] = df['release_year'].apply(lambda x: x+2)
```

```python
df.head()
```

```python
df['date_added'] = df['date_added'].apply(lambda x: 'Pre Covid' if x<pd.Timestamp('2020-04-01') else 'Post Covid')
```

```python
df.head()
df.tail()
df.sample(10)
```

```python
# we can also transform like this by numpy vectorisation using `where` function 



data = pd.read_csv('Dataset\Orginal Netflix.csv')

data.head()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], format='mixed')
data.head()
```

```python
# pre covid : before 2018-04-01 & after - post covid 

# using numpy where clause 


data['Covid stat'] = np.where(data['date_added']<pd.Timestamp('2018-04-01'), 'Pre Covid', 'Post Covid')
```

```python
data.head()
```
